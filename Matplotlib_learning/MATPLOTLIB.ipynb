### Which method is used ? i.e. (pyplot vs matplotlib oo method)

* when plotting something quickly , then you can use pyplote methods.
* But plotting something importing or more advance then use the OO Method.

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
## Pyplote method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
#3 oo method
import matplotlib.pyplot as plt
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
## method 2 of oo method

fig, ax=plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Cuatomize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle= '--');
```

```python
## Subplots of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data at ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
        ylabel="Cholesterol");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= '--');




# Add data at ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# customize ax1
ax1.set(title="Heart disease and Max Heart Rate",
        xlabel= "Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle= '--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

## method 2 of oo method

fig, ax=plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap="winter") # this will change the coluri scheme

# Cuatomize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle= '--');
```

```python
## Customizing the y and x axis limitations

## Subplots of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

# Add data at ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
        ylabel="Cholesterol");

## Change the x axix limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle= '--');




# Add data at ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

# customize ax1
ax1.set(title="Heart disease and Max Heart Rate",
        xlabel= "Age",
        ylabel="Max Heart Rate")

## Change the x axix limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle= '--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
