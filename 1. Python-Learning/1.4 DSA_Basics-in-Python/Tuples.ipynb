## Tuples

Tuples are ordered collections of items that are immutable . They are similar to list, but their immutability makes them different.

```python
## Creating a Tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6,7,8,9])
print(numbers)
## This means that the list can be converted into tuple.
```

```python
print(type(numbers))
```

```python
list((1,2,3,4,5,6,7,8))
## this is the another way to display the list
```

```python
mixed_tuple=(1,"hello world",3.14,True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements
numbers
```

```python
numbers[2]
```

```python
print(numbers[-1])
## It will pickup the last index 
```

```python
## Slicing Of Tuple
numbers[0:4]
```

```python
print(numbers[:4])
print(numbers[::-1])
```

```python
## Tuple Operations

concatenation_tuple=numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
## In this the entire tuple is getting appended 3 times
```

```python
numbers * 3
## Here also the same thing will happen.
```

```python
## Immutable nature of Tuples

## 1. Tuples are immutable, meaning their elements cannot be changed once assigned.
lst=[1,2,3,4]
print(lst)

lst[1]="Rishi"
print(lst)

## As you can see we can change the values in list . But we can't do in tuples.
## Lets see the exampels

```

```python
numbers[1]="Rishi"

## As you can see you cannot change the tuple values.
## There are ways to changes to change the values by converting the
## tuple into the list and then we can change the tuple values
## and then you can change the list back into the tuple.
```

```python
# You can do this by these steps

# Original tuple
numbers = (1, 2, 3, 4)  # Assuming this is your tuple

# Convert tuple to list
numbers_list = list(numbers)

# Modify the list
numbers_list[1] = "Rishi"

# Convert list back to tuple
numbers = tuple(numbers_list)

print(numbers)  # Output: (1, 'Rishi', 3, 4)

## As you can see you cannot change the tuple values directly.
## This is the way to change the values by converting the
## tuple into the list and then we can change the values
## and then convert the list back into the tuple.
```

```python
## Tuples Methods
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking Tuples
## Packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"hello",3.14,"c"]]
lst[0]
```

```python
lst[0][2]
```

```python
lst[0][0:3]
## slicing operation can also be done in nested list
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"hello",3.14,"c")]
lst[0][0:3]
## Inside a list you can have tuples as well.
```

```python
nested_tuples=((1,2,3),("a","b","c"),(True,False))
print(nested_tuples[0])
print(nested_tuples[1][2])

## Access the elements inside a tuple
```

```python
## Iterating over nested tuples
for sub_tuple in nested_tuples:
    for item in sub_tuple:
        print(item,end=" ")
    print()    
```
