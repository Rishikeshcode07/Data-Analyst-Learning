##### Dictionaries

Dictionaries are unorderd collections of items. They atore data in key-value pairs.Keys must be unique and immutable(eg - strings,numbers or tuples) while values can be of any type.

```python
## Creating Dictionaries

empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

So these are the two ways you can create empty dictionaries

```python
student={"name":"Rishi","age":32,"grade":24}
print(student)
```

```python
## When you get error
student={"name":"Rishi","age":32,"name":24}
print(student)

## So as you can see the values are less printed because of the key (which is used two times) 
```

```python
## Accessing dictionaries elements

student={"name":"Rishi","age":32,"grade":'A'}
print(student)

print(student['grade'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifing Dictionarie Elements
## Dictionary are mutaable, os you can add,update or delete elements

print(student)

student["age"]=33  ## update value for the key
print(student)

student["address"]="India"  ## added a new key and value
print(student)

del student['grade']   ## delete key and value pair
print(student)
```

```python
## Dictionary methods

keys=student.keys()  ## get all the keys
print(keys)
values=student.values()  ## get all the values
print(values)


items=student.items()  ## get all key value pairs
print(items)
```

```python
## Shallow copy
## It provides different memeory allocations
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Rishi2"
print(student_copy)

student_copy = student.copy()  ## shallow copy
print(student_copy)
print(student)
```

```python
### Iteration over dictionaries
## You can use loops to iterate over dictionatris, keys, values , or items


## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for values in student.values():
    print(values)
```

```python
#3 Iterate over key value paris

for key,value in student.items():
    print(f"{keys}:{value}")
```

```python
## Nested Dictionaries

students={
    "student1":{"name":"rishi","age":32},
    "student2":{"name":"rahul","age":30}
}
print(students)
```

```python
## Access nested dictionareis
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iteration over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")

```

```python
## Dictionary Comphrehension with conditions
squares={x:x**2   for x in range(5)}
print(squares)

evens={x:x**2  for x in range(10)  if x%2==0}
print(evens)
```
