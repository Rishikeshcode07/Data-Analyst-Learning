### File Operation- Read And Write Files

File handling is a crucial part of any programming language. Python provides built-in function and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling , including reading and writing text files and binary files.

```python
### Read a whole file
### Heera 'r' means read a file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
### Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line)
        ## This line will print the output with a extra line
```

```python
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) 
        ## By using this function it will print the output without an extra line
```

```python
### Writing a file (OverWriting)

with open('example.txt','w') as file:
    file.write('Hello World its me. :\n')
    file.write('This is a new line.')
## Just by using this the output will be overwritten in the file i.e the old lines will be fanished.
    
```

```python
### Write a file without overwriting
## This 'a' means append which means the old file will be get executed long with the new file.

with open('example.txt','a') as file:
    file.write('Append operation taking place!\n')
    
```

```python
### Writing a list of lines to a file

lines=['First line \n','Second Line \n','Third Line \n']
with open('example.txt','a') as file:
    file.writelines(lines)  ## This function will add the no of lines in the file
     
```

```python
### Binary Files

# Writing to a binary file
data= b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
    print(data)

## This 'wb' is write bit 
```

```python
data= b'\x00\x01\x02\x03\x04'
with open('example.bin','rb') as file:
    content=file.read()
    print(content)

## This 'rb' is read bit 
```

```python
## Read the content from a source text file and write to a destination text file
# Copying a text file

with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)  


''' 
Here you can see that , i have not creeated the destination file it 
has been automaticaly created by the system.

So this ishow you can copy a file and past to another file without creating
a new file which will be created by own.
'''      
```

```python
## Read a text file and count the number of lines, words, and characters.
## Count lines,words, and characters in a text file

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines= file.readlines()
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines) 
        char_count= sum(len(line) for line in lines)
    return line_count,word_count,char_count

file_path ='example.txt'
lines,words,characters= count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')    
```
