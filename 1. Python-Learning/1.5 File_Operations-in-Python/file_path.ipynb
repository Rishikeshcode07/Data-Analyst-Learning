```python
### Usig the os module

import os
cwd = os.getcwd()
print(f"Current working directory is {cwd}")
```

```python
## Create a new directory

new_directory="package"
os.mkdir(new_directory)
print(f"Directory '{new_directory}' created")
```

```python
### List files and directories
items = os.listdir('.')
print(items)
```

```python
### Joining paths

dir_name="folder"
file_name="file.txt"
full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
## Getting a file along with folder location
dir_name="folder"
filn_name="file.txt"
full_path=os.path.join(os.getcwd(),dir_name,filn_name)
print(full_path)
```

```python
## Check wheather the file exists or not
## 1. method
path='example.txt'
if os.path.exists(path):
    print(f"The path '{path}' exists")
else:
    print(f"The path '{path}' does not exists")    
```

```python
## 2. method

import os

path='example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory") 
else:
    print(f"The path '{path}' is a neither file nor a directory")       
```

```python
## Getting the absulote path
relative_path= 'example.txt'
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
