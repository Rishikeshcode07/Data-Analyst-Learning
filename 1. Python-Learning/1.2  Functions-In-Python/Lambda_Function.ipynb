#### Lambda Functions In Pyhton

Lambda functions are small anonymous functions(function without a name) defined using the lambda keywords. They can have any number of argumnets but only one expression(one logic only). They are commonly used for short or as arguments to higher-order functions.


```python
## Syntax
lambda argumnets: expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(2,3)
```

```python
## So as you can see that just for one expression you have created 
## the entire function for this . So what the lambda function does is that 
## It basically convert that function into single expression so that you 
## don't have to create such a big function.

## Lets see how to convert that function into just single line 

addition=lambda a,b:a+b  ## this will convert your function into single one
type(addition)


```

```python
print(addition(3,4)) ## You can see the output
```

```python
## Lets take another example
## First the basic function
def even(num):
    if num%2==0:
        return True
even(34)    
```

```python
even_number=lambda n:n%2==0
print(even_number(4))
```

```python
addition_number=lambda x,y,z:x+y+z
addition_number(12,24,56)
```
