Definatons:- 

A function is a block of code that performs a specific task. Functions help in organizing code, resuing code, and improveing readability.

```python
#3 Syntax

def function_name(parameters):
    """Docstring"""  ## It helps in writing some comments of what the function does

    # Function body
    return expression
```

```python
## Why functions?

def even_or_odd(num):
    """This function finds even or odd"""

    if num%2==0:
        print(" The number is even")
    else:
        print("The number is odd")    
```

```python
## Call a funtion
even_or_odd(33)
```

```python
## Function with multiple parameters
def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
## Default Parameters
def greet(name):
    print(f"hello {name} welcome to the paradise")

greet("Rishi")
    
```

```python
def greet(name="Guest"):
    print(f"hello {name} welcome to the paradise")

greet()
## IN this if you don't pass the parameter then the default value will be passed
 
```

```python
## Variable Length Argumments
## Positional and keywords argumnents

## Positional argumnets are those argumnets that are passed in the 
## in the actual value time. Here the '*' means all 

def print_numbers(*Rishi):
    for number in Rishi:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,"Rishi") ## Positional argument
```

```python
## Positional arguments
## This is the best practise to write the positional arguments

def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,"Rishi")
```

```python
## Keywords arguments
## In keyword arguments all the values are in the form of
## "key:values" pair

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
        
```

```python
print_details(name="Rishi",age="32",country="India")
```

```python
## Combining both

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Rishi",name="Rishi",age="32",country="India")
```

```python
## Return Multiple Parameters
def multiply(a,b):
    return a*b,a

multiply(2,3)
```

### Few examples to understand the function in more better way


```python
## Example 1
## Temperature Conversion

def convert_temperature(temp,unit):
    """This function converts temperature between Celsius and Fahrenheit"""

    if unit =='C':
        return temp * 9/5 + 32
    elif unit == "F":
        return (temp-32)*5/9
    else:
        return None
    
print(convert_temperature(25,'C'))
print(convert_temperature(77,'F'))


```

```python
## Example 2
## Password Strength Checker

def is_strong_password(password):
    """This function checks if the password is strong or not"""

    if len(password)<8:
        return False
    if not any(char.isdigit()    for char in password):
        return False
    if not any(char.islower()   for char in password):
        return False
    if not any(char.isupper()   for char in password):
        return False
    if not any(char in '!@#$%^&*_+-()'   for char in password):
        return False
    return True

## Calling the function
print(is_strong_password("Weakpwd"))
print(is_strong_password("Rishikesh@1"))
```

```python
## CALCULATE THE TOTAL COST OF ITEMS IN A SHOPPING CART

def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']* item['quantity']

    return total_cost    


cart=[
    {'namae':'Apple','price':0.5,'quantity':4},
    {'namae':'Banana','price':0.3,'quantity':6},
    {'namae':'Orange','price':0.7,'quantity':3},
]

total_cost=calculate_total_cost(cart)
print(total_cost)
```

```python
## CHECK IF A STRING IS PALINDROME

def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("Hello sir"))
```

```python
## CALCULATE THE FACTORIAL OF A NUMBER USING RECURSION

def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

print(factorial(5))    
```

```python
## A FUNCTION TO READ A FILE AND COUNT THE FREQUENCY OF EACH WORD

def count_word_frequency(file_path):
    word_count={}
    with open(file_path,'r') as file:
        for line in file:
            words=line.split()
            for word in words:
                word= word.lower().strip('.,!?:;"\'')
                word_count[word]=word_count.get(word,0)+1
    return word_count  

filepath ='sample.txt' 
word_frequency=count_word_frequency(filepath)   
print(word_frequency)     
```
