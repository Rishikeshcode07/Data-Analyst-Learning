# WHAT IS PANDAS ?

Pandas is a tool (a library) in Python that helps you work with data, like numbers, text, or dates, in an easy way. Think of it like a super smart spreadsheet (like Excel) that you can control with code. It organizes data into tables (called DataFrames) or lists (called Series), so you can sort, filter, or analyze it.



## WHY PANDAS IS USED ?

Pandas is used because it makes working with data easier and faster. Here’s why people love it:

* Organizes Data: It puts messy data into neat tables, so it’s easy to understand and work with.

  
* Cleans Data: It can fix problems like missing information or duplicate entries.

* Analyzes Data: You can quickly calculate things like averages, totals, or group data (e.g., "show total sales by city").

  
* Works with Files: It can read data from files like CSV or Excel and save your results back to them.

  
* Saves Time: It’s fast and simple to use, so you don’t need to write lots of complicated code.


* Works with Other Tools: It teams up with other Python tools for making charts or building machine learning models.

```python
car_sales_missing
```

```python
''' 
But if you can see there is still no values in the nan place
so to deal with that we have something called 
inplace.
'''

car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),inplace=True)
car_sales_missing
```
